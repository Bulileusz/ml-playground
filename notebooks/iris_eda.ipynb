```python
# --- 1. Importy ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# --- 2. Wczytanie danych ---
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# --- 3. Podstawowe informacje ---
print("Rozmiar danych:", df.shape)
print("\nPierwsze wiersze:")
display(df.head())

print("\nTypy danych:")
print(df.dtypes)

print("\nStatystyki opisowe:")
display(df.describe())

# --- 4. Sprawdzenie braków danych ---
print("\nBrakujące wartości:")
print(df.isnull().sum())

# --- 5. Wizualizacje ---
sns.pairplot(df, hue='species', diag_kind='hist')
plt.suptitle("Para-wykresy cech Iris", y=1.02)
plt.show()

plt.figure(figsize=(8, 5))
sns.heatmap(df.drop(columns='species').corr(), annot=True, cmap='coolwarm')
plt.title("Macierz korelacji cech")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x='species', y=df.columns[0], data=df)
plt.title(f"Rozkład cechy '{df.columns[0]}' dla poszczególnych gatunków")
plt.show()
```
